**Real World Example: Loan Calculator**

```python
# Real World Example: Loan Calculator
print("\n--- Real Example: Loan Calculator ---")
```

```python
def calculate_monthly_payment(loan_amount, annual_rate, years):
    """Calculate monthly loan payment"""
    monthly_rate = annual_rate / 100 / 12
    num_payments = years * 12
    
    if monthly_rate == 0:
        monthly_payment = loan_amount / num_payments
    else:
        monthly_payment = loan_amount * (monthly_rate * (1 + monthly_rate)**num_payments) / \
                         ((1 + monthly_rate)**num_payments - 1)
    
    total_payment = monthly_payment * num_payments
    total_interest = total_payment - loan_amount
    
    return monthly_payment, total_payment, total_interest
```

```python
# Calculate for a $200,000 loan
loan = 200000
rate = 3.5
term = 30
monthly, total, interest = calculate_monthly_payment(loan, rate, term)
```

```python
print(f"Loan Amount: ${loan:,}")
print(f"Interest Rate: {rate}%")
print(f"Term: {term} years")
print(f"\nðŸ’° Monthly Payment: ${monthly:,.2f}")
print(f"ðŸ’° Total Payment: ${total:,.2f}")
print(f"ðŸ’° Total Interest: ${interest:,.2f}")
```

**List Comprehensions**

```python
# Traditional way vs List Comprehension
print("\n--- Traditional Way ---")
squares_traditional = []
for i in range(1, 6):
    squares_traditional.append(i ** 2)
print(f"Squares (traditional): {squares_traditional}")
```

```python
print("\n--- List Comprehension Way ---")
squares = [i ** 2 for i in range(1, 6)]
print(f"Squares (comprehension): {squares}")
```

```python
# List comprehension with condition
print("\n--- With Condition ---")
even_numbers = [i for i in range(1, 11) if i % 2 == 0]
print(f"Even numbers: {even_numbers}")

odd_numbers = [i for i in range(1, 11) if i % 2 != 0]
print(f"Odd numbers: {odd_numbers}")
```

```python
# String operations with list comprehension
print("\n--- String Operations ---")
names = ['john', 'sarah', 'mike', 'emma']
capitalized = [name.title() for name in names]
print(f"Original: {names}")
print(f"Capitalized: {capitalized}")
```

```python
# List comprehension with multiple conditions
print("\n--- Multiple Conditions ---")
numbers = [i for i in range(1, 51) if i % 3 == 0 and i % 5 == 0]
print(f"Numbers divisible by both 3 and 5 (1-50): {numbers}")
```

```python
# Nested list comprehension
print("\n--- Nested Comprehension ---")
matrix = [[i*j for j in range(1, 4)] for i in range(1, 4)]
print("Multiplication table:")
for row in matrix:
    print(row)
```

**Real World Example: Data Cleaning**

```python
# Real World Example: Data Cleaning
print("\n--- Real Example: Price Data Cleaning ---")
raw_prices = ['$100', '$250', '$75', '$350', '$125']
print(f"Raw prices: {raw_prices}")
```

```python
# Clean and convert to integers
clean_prices = [int(price.replace('$','')) for price in raw_prices]
print(f"Clean prices: {clean_prices}")
```

```python
# Filter expensive items (over $100)
expensive_items = [price for price in clean_prices if price > 100]
print(f"Expensive items (>$100): {expensive_items}")
```

```python
# Apply discount (20% off)
discounted_prices = [price * 0.8 for price in clean_prices]
print(f"After 20% discount: {discounted_prices}")
```

**Real World Example: Customer Data Processing**

```python
# Real World Example: Customer Data Processing
print("\n--- Real Example: Customer Data Processing ---")
customers = [
    {'name': 'John', 'age': 25, 'purchases': 150},
    {'name': 'Sarah', 'age': 32, 'purchases': 420},
    {'name': 'Mike', 'age': 28, 'purchases': 280},
    {'name': 'Emma', 'age': 35, 'purchases': 550},
]
```

```python
# Extract names
names = [customer['name'] for customer in customers]
print(f"Customer names: {names}")
```

```python
# Find VIP customers (purchases > $300)
vip_customers = [customer['name'] for customer in customers 
                 if customer['purchases'] > 300]
print(f"VIP customers: {vip_customers}")
```

```python
# Calculate total revenue
total_revenue = sum([customer['purchases'] for customer in customers])
print(f"Total revenue: ${total_revenue}")
```
